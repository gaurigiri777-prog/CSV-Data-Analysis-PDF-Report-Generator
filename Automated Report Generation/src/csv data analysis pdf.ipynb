```python
import pandas as pd
import matplotlib.pyplot as plt
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

# -------------------------------
# Step 1: Load Data
# -------------------------------
def load_data(file_path):
    df = pd.read_csv(file_path, encoding='latin1')
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    print("✅ Data loaded successfully!")
    return df


# -------------------------------
# Step 2: Analyze Data
# -------------------------------
def analyze_data(df):
    summary = {}

    summary['total_orders'] = len(df)
    summary['total_sales'] = df['Sales'].sum()
    summary['total_profit'] = df['Profit'].sum()
    summary['average_sales'] = df['Sales'].mean()

    summary['sales_by_category'] = df.groupby('Category')['Sales'].sum()
    summary['sales_by_region'] = df.groupby('Region')['Sales'].sum()

    # Monthly Trend
    df['Month'] = df['Order Date'].dt.to_period('M')
    summary['monthly_sales'] = df.groupby('Month')['Sales'].sum()

    summary['top_category'] = summary['sales_by_category'].idxmax()
    summary['top_region'] = summary['sales_by_region'].idxmax()

    return summary


# -------------------------------
# Step 3: Create Charts
# -------------------------------
def create_charts(summary):
    chart_files = []

    # Chart 1: Sales by Category
    plt.figure()
    summary['sales_by_category'].plot(kind='bar')
    plt.title("Sales by Category")
    plt.xlabel("Category")
    plt.ylabel("Sales")
    file1 = "category_chart.png"
    plt.savefig(file1)
    plt.close()
    chart_files.append(file1)

    # Chart 2: Sales by Region
    plt.figure()
    summary['sales_by_region'].plot(kind='bar')
    plt.title("Sales by Region")
    plt.xlabel("Region")
    plt.ylabel("Sales")
    file2 = "region_chart.png"
    plt.savefig(file2)
    plt.close()
    chart_files.append(file2)

    # Chart 3: Monthly Sales Trend
    plt.figure()
    summary['monthly_sales'].plot()
    plt.title("Monthly Sales Trend")
    plt.xlabel("Month")
    plt.ylabel("Sales")
    file3 = "monthly_chart.png"
    plt.savefig(file3)
    plt.close()
    chart_files.append(file3)

    return chart_files


# -------------------------------
# Step 4: Generate PDF
# -------------------------------
def create_pdf(summary, df, chart_files, output_file):
    doc = SimpleDocTemplate(output_file)
    styles = getSampleStyleSheet()
    elements = []

    # Title
    elements.append(Paragraph("Superstore Sales Analysis Report", styles['Title']))
    elements.append(Spacer(1, 20))

    # Summary
    elements.append(Paragraph("📊 Key Insights", styles['Heading2']))
    elements.append(Paragraph(f"Total Orders: {summary['total_orders']}", styles['Normal']))
    elements.append(Paragraph(f"Total Sales: {summary['total_sales']:.2f}", styles['Normal']))
    elements.append(Paragraph(f"Total Profit: {summary['total_profit']:.2f}", styles['Normal']))
    elements.append(Paragraph(f"Average Sales: {summary['average_sales']:.2f}", styles['Normal']))
    elements.append(Paragraph(f"Top Category: {summary['top_category']}", styles['Normal']))
    elements.append(Paragraph(f"Top Region: {summary['top_region']}", styles['Normal']))
    elements.append(Spacer(1, 20))

    # Charts Section
    elements.append(Paragraph("📈 Visual Insights", styles['Heading2']))
    elements.append(Spacer(1, 10))

    for chart in chart_files:
        elements.append(Image(chart, width=400, height=250))
        elements.append(Spacer(1, 20))

    # Sample Data Table
    elements.append(Paragraph("📄 Sample Data", styles['Heading2']))

    table_data = [df.columns.tolist()] + df.head(10).values.tolist()

    table = Table(table_data)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('GRID', (0, 0), (-1, -1), 0.5, colors.black),
    ]))

    elements.append(table)

    doc.build(elements)
    print("✅ PDF with charts created successfully!")


# -------------------------------
# Main Execution
# -------------------------------
if __name__ == "__main__":
    file_path = r"C:\Users\gauri\OneDrive\Desktop\ML\dataset\Sample - Superstore.csv"
    output_file = "superstore_report_with_charts.pdf"

    df = load_data(file_path)
    summary = analyze_data(df)
    chart_files = create_charts(summary)
    create_pdf(summary, df, chart_files, output_file)
```
